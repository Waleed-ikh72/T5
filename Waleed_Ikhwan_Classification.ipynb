```python

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import pandas as pd
# Loading the iris dataset
iris = datasets.load_iris()
iris

#
```

```python
# Creating a dataframe
data=pd.DataFrame({
    'sepal length':iris.data[:,0],
    'sepal width':iris.data[:,1],
    'petal length':iris.data[:,2],
    'petal width':iris.data[:,3],
    'species':iris.target
})
data
```

```python

import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
data.isna().sum()
```

```python

# Splitting the dataset into training and testing sets
X = data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y = data['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Creating a random forest classifier
clf=RandomForestClassifier(n_estimators=100)

# Training the classifier
clf.fit(X_train,y_train)

# Predicting the species for the testing set
y_pred=clf.predict(X_test)

# Evaluating the accuracy of the classifier
print("Accuracy:%",metrics.accuracy_score(y_test, y_pred)*100)
```

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.linear_model import LogisticRegression
# Plotting the decision boundary of sepal length and sepal width
pair = [0, 1]
X = data.iloc[:, pair].values
y = data['species'].values

# Train a logistic regression model
clf = LogisticRegression().fit(X, y)

# Plot the decision boundary
_, ax = plt.subplots()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    cmap=plt.cm.Paired,
    response_method="predict",
    ax=ax,
)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors="k")
plt.xlabel(data.columns[pair[0]])
plt.ylabel(data.columns[pair[1]])
plt.title("Decision Boundary of Sepal Length and Sepal Width")
plt.show()
```
