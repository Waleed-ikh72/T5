```python
# Handling Missing values For numerical columns

numerical_columns = df.select_dtypes(include='float64')
numerical_columns
```

```python
# plot numerical values to see if they are normally distributed or not

for col in numerical_columns:
  sns.histplot(df[col],kde = True)
  plt.xlabel(col)
  plt.title(f'Distribution of {col}')
  plt.show()
```

```python
df.isna().sum()
```

```python
# We will impute the missing values with mean since it is normal distributions

for col in numerical_columns:
  df[col].fillna(df[col].mean(),inplace = True)
```

```python
df.isna().sum()
```

```python
# Handking missing values in categrical columns

categorical_columns = df.select_dtypes(include='object')
categorical_columns
```

```python
for col2 in categorical_columns:
  df.dropna(subset=[col2],inplace = True)
```

```python
df.reset_index(inplace=True, drop=True)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Encode categorical features
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Job Title'] = label_encoder.fit_transform(df['Job Title'])
df['Education Level'] = label_encoder.fit_transform(df['Education Level'])

# Separate features and target variable
X = df.drop('Salary', axis=1)
y = df['Salary']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale numerical features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```
